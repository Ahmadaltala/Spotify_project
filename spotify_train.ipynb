```python
# Load the dataset
data = pd.read_csv('Train.csv')
```

```python
# Display basic information about the dataset
print("Basic Information About the Dataset:")
data.info()
```

```python
# Preview the dataset
print("\nPreview of the Dataset:")
print(data.head())
```

```python
# Check for missing values
print("\nMissing Values in Each Column:")
print(data.isnull().sum())
```

```python
# Handle missing values in 'top genre' by filling with 'Unknown'
data['top genre'] = data['top genre'].fillna('Unknown')
```

```python
# Select only numerical columns for the correlation matrix
numerical_data = data.select_dtypes(include=['number'])
```

```python
# Correlation heatmap for numerical features
plt.figure(figsize=(10, 8))
correlation_matrix = numerical_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot for a single feature vs. popularity (e.g., energy vs. popularity)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=data['nrgy'], y=data['pop'])
plt.title('Energy vs Popularity')
plt.xlabel('Energy (nrgy)')
plt.ylabel('Popularity (pop)')
plt.show()
```

```python
# Histogram of the target variable 'pop'
plt.figure(figsize=(8, 5))
sns.histplot(data['pop'], bins=20, kde=True, color='blue')
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```
